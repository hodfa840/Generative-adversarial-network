```python
import numpy as np
import matplotlib.pyplot as plt
# from sklearn.neighbors import KernelDensity
from scipy.stats import gaussian_kde
from scipy.stats import norm
from scipy.integrate import quad
```

```python
# load numpy array from npy file
from numpy import load
# load array
model4 = load('model4.npy')
real4 = model4[0,:]
fake4 = model4[1,:]
```

```python
xmin = np.min([real4,fake4])
xmax = np.max([real4,fake4])
```

```python
n_bins = 50
bins = np.linspace(xmin, xmax, n_bins)
h1, edges = np.histogram(fake4, bins=bins)
h2, edges = np.histogram(real4, bins=bins)

h1 = h1 / np.sum(h1)
h2 = h2 / np.sum(h2)
```

```python
plt.bar(edges[:-1], h1, alpha = 0.5, width = 0.2,label = 'fake')
plt.bar(edges[:-1], h2, alpha = 0.5, width = 0.2,label = 'real')
plt.legend()
plt.show()
```

```python
sumover4 =  h1 * np.log((h1 + 1e-6) / (h2 + 1e-6))

plt.bar(edges[:-1], sumover4,width = 0.2)
# plt.bar(edges[:-1], h1, alpha = 0.5, width = 0.2)
# plt.bar(edges[:-1], h2, alpha = 0.5, width = 0.2)
plt.show()
```

```python
discrete_kl4 = np.sum(sumover4)
discrete_kl4
```

```python
dis1  = fake4
dist2 = real4
x = np.linspace(xmin, xmax, 100)
```

```python
n_samples = 10000
draws1 = fake4.squeeze()
draws2 = real4.squeeze()

# bw_method = .5
bw_method = 'scott'
kde1 = gaussian_kde(draws1, bw_method=bw_method)
kde2 = gaussian_kde(draws2, bw_method=bw_method)

plt.hist([draws1, draws2], bins=15, density=True)
plt.plot(x, kde1(x))
plt.plot(x, kde2(x))
```

```python
f = lambda x: kde1(x) * np.log((kde1(x) + 1e-6) / (kde2(x) + 1e-6))

plt.plot(x, f(x))
plt.show()
```

```python
kernel_integral4, error = quad(f, xmin, xmax)
kernel_integral4, error
```

```python
print("Numerical integration of kernels\t", kernel_integral4)
print("Discretization using histograms\t\t", discrete_kl4)
```
